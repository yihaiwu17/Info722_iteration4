```python
from pyspark.sql.window import Window
from pyspark.sql import functions as F
import matplotlib.pyplot as plt

# Calculate the median number of samples per class
n_samples_per_class = df.groupBy('Country').count().approxQuantile('count', [0.5], 0.01)[0]

# Create a balanced DataFrame
resampled_list = []

for country in df.select('Country').distinct().collect():
    country_name = country['Country']
    group = df.filter(col('Country') == country_name)
    resampled_class = group.sample(withReplacement=True, fraction=(n_samples_per_class / group.count()))
    resampled_list.append(resampled_class)

# Concatenate all resampled slices back into a single DataFrame
df_balanced = resampled_list[0]
for df_part in resampled_list[1:]:
    df_balanced = df_balanced.union(df_part)

# Shuffle the DataFrame
df_balanced = df_balanced.orderBy(F.rand())

# Verify the balancing
df_balanced.groupBy('Country').count().show()
```

```python
# Plotting the balanced data
country_counts = df_balanced.groupBy('Country').count().toPandas()

# Bar chart
plt.figure(figsize=(16, 10))
plt.bar(country_counts['Country'], country_counts['count'], color='skyblue')
plt.xlabel('Country', fontsize=15)
plt.ylabel('Count', fontsize=15)
plt.title('Balanced Country Distribution', fontsize=20)
plt.xticks(rotation=45, ha='right', fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```

```python
from pyspark.ml.regression import LinearRegression, DecisionTreeRegressor
from pyspark.ml.evaluation import RegressionEvaluator
from pyspark.ml import Pipeline

# Split the data into training and testing sets
train_data, test_data = df.randomSplit([0.8, 0.2], seed=42)

# Initialize the models
linear_model = LinearRegression(featuresCol='features', labelCol='PercentageValue')
ridge_model = LinearRegression(featuresCol='features', labelCol='PercentageValue', regParam=1.0)  # Ridge regression with regularization
decision_tree_model = DecisionTreeRegressor(featuresCol='features', labelCol='PercentageValue')

# Pipelines
linear_pipeline = Pipeline(stages=indexers + encoders + [assembler, linear_model])
ridge_pipeline = Pipeline(stages=indexers + encoders + [assembler, ridge_model])
decision_tree_pipeline = Pipeline(stages=indexers + encoders + [assembler, decision_tree_model])

# Train the models
linear_model_fitted = linear_pipeline.fit(train_data)
ridge_model_fitted = ridge_pipeline.fit(train_data)
decision_tree_model_fitted = decision_tree_pipeline.fit(train_data)

# Make predictions
linear_predictions = linear_model_fitted.transform(test_data)
ridge_predictions = ridge_model_fitted.transform(test_data)
decision_tree_predictions = decision_tree_model_fitted.transform(test_data)

# Initialize the evaluator
evaluator = RegressionEvaluator(labelCol="PercentageValue", predictionCol="prediction", metricName="r2")

# Calculate R-squared
linear_r2 = evaluator.evaluate(linear_predictions)
ridge_r2 = evaluator.evaluate(ridge_predictions)
decision_tree_r2 = evaluator.evaluate(decision_tree_predictions)

print(f"Linear Regression R^2: {linear_r2}")
print(f"Ridge Regression R^2: {ridge_r2}")
print(f"Decision Tree Regression R^2: {decision_tree_r2}")
```

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import r2_score, mean_squared_error
import numpy as np
import pandas as pd

# Define a pipeline for preprocessing
preprocess_pipeline = Pipeline(stages=indexers + encoders + [assembler])

# Fit and transform the training data
preprocess_model = preprocess_pipeline.fit(train_data)
train_data_transformed = preprocess_model.transform(train_data)
test_data_transformed = preprocess_model.transform(test_data)

# Convert to Pandas DataFrame
train_pd = train_data_transformed.select('features', 'PercentageValue').toPandas()
test_pd = test_data_transformed.select('features', 'PercentageValue').toPandas()

# Extract features and labels
X_train = np.array(train_pd['features'].tolist())
y_train = train_pd['PercentageValue'].values
X_test = np.array(test_pd['features'].tolist())
y_test = test_pd['PercentageValue'].values

# Train KNN Regressor using scikit-learn
knn_model = KNeighborsRegressor(n_neighbors=5)
knn_model.fit(X_train, y_train)

# Make predictions
knn_predictions = knn_model.predict(X_test)

# Calculate R-squared and RMSE
knn_r2 = r2_score(y_test, knn_predictions)
knn_rmse = mean_squared_error(y_test, knn_predictions, squared=False)

print(f"KNN Regressor R^2: {knn_r2}")
print(f"KNN Regressor RMSE: {knn_rmse}")
```

```python
# Assembling features
assembler = VectorAssembler(inputCols=['Variable_vec', 'Country_vec', 'Year'], outputCol="features")

# Split the data into training and testing sets
train_data, test_data = df.randomSplit([0.8, 0.2], seed=42)

# Configure the Decision Tree Regressor with specific parameters
decision_tree_model = DecisionTreeRegressor(
    featuresCol='features',
    labelCol='PercentageValue',
    maxDepth=10,
    maxBins=32,
    minInstancesPerNode=2,
    minInfoGain=0.0,
    maxMemoryInMB=256,
    cacheNodeIds=False,
    checkpointInterval=10,
    impurity='variance'
)

# Pipeline
pipeline = Pipeline(stages=indexers + encoders + [assembler, decision_tree_model])

# Train the model
model = pipeline.fit(train_data)

# Predict on the test data
predictions = model.transform(test_data)

# Initialize the evaluator
evaluator = RegressionEvaluator(labelCol="PercentageValue", predictionCol="prediction")

# Calculate R-squared
r2 = evaluator.evaluate(predictions, {evaluator.metricName: "r2"})
print(f"R-squared on test data: {r2}")

# Extract the trained decision tree model
tree_model = model.stages[-1]

# Print the tree structure
tree_structure = tree_model.toDebugString
print(tree_structure)
```

```python
# Prepare the feature names based on the one-hot encoding
variable_feature_names = [f"Variable_{name}" for name in indexers[0].fit(df).labels]
country_feature_names = [f"Country_{name}" for name in indexers[1].fit(df).labels]
feature_names = variable_feature_names + country_feature_names + ['Year']

# Create a mapping from feature index to feature name
feature_index_to_name = {i: name for i, name in enumerate(feature_names)}

```

```python
import re

# Function to replace feature indices with feature names in the tree structure
def replace_feature_indices(tree_structure, feature_index_to_name):
    def replacer(match):
        index = int(match.group(1))
        return feature_index_to_name.get(index, f"feature {index}")
    
    pattern = re.compile(r'feature (\d+)')
    replaced_structure = pattern.sub(replacer, tree_structure)
    return replaced_structure

# Replace feature indices in the tree structure
tree_structure_with_names = replace_feature_indices(tree_structure, feature_index_to_name)
print(tree_structure_with_names)
```

```python
from pyspark.ml.tuning import CrossValidator, ParamGridBuilder
from pyspark.ml.evaluation import RegressionEvaluator

# Prepare the Decision Tree Regressor
decision_tree = DecisionTreeRegressor(featuresCol='features', labelCol='PercentageValue')

# Create a parameter grid for cross-validation
paramGrid = ParamGridBuilder() \
    .addGrid(decision_tree.maxDepth, list(range(1, 11))) \
    .build()

# Initialize the evaluator
evaluator = RegressionEvaluator(labelCol="PercentageValue", predictionCol="prediction", metricName="rmse")

# Set up CrossValidator
crossval = CrossValidator(estimator=decision_tree,
                          estimatorParamMaps=paramGrid,
                          evaluator=evaluator,
                          numFolds=10)  # 10-fold cross-validation

# Pipeline with the cross-validation
pipeline_cv = Pipeline(stages=indexers + encoders + [assembler, crossval])

# Run cross-validation and choose the best set of parameters
cv_model = pipeline_cv.fit(train_data)

# Extract the best model
best_model = cv_model.stages[-1].bestModel

# Print the optimal depth
optimal_depth = best_model.getOrDefault('maxDepth')
print(f"Optimal depth: {optimal_depth}")

# Calculate and print RMSE for each depth
average_rmse = cv_model.stages[-1].avgMetrics
print(f"Average RMSE for each depth: {average_rmse}")

```
